```python
import ipywidgets as widgets
from IPython.display import display

slider_layout = widgets.Layout(width='90%')  # wider sliders

# Sliders with larger width and font size
t_infall_slider = widgets.FloatRangeSlider(
    value=[0, 13.82], min=0, max=15, step=0.01, 
    description='t_infall', layout=slider_layout, style={'description_width': '150px'}
)

r_peri_slider = widgets.FloatRangeSlider(
    value=[-10, 200], min=-50, max=250, step=1, 
    description='r_peri', layout=slider_layout, style={'description_width': '150px'}
)

r_apo_slider = widgets.FloatRangeSlider(
    value=[-10, 400], min=-50, max=450, step=1, 
    description='r_apo', layout=slider_layout, style={'description_width': '150px'}
)

angle_slider = widgets.FloatRangeSlider(
    value=[-5, 95], min=-10, max=100, step=1, 
    description='angle', layout=slider_layout, style={'description_width': '150px'}
)

sigma_slider = widgets.FloatRangeSlider(
    value=[2, 150], min=1, max=200, step=1, 
    description='sigma', layout=slider_layout, style={'description_width': '150px'}
)

r50_slider = widgets.FloatRangeSlider(
    value=[0.011, 3], min=0, max=5, step=0.01, 
    description='r50', layout=slider_layout, style={'description_width': '150px'}
)

m_stars_slider = widgets.FloatRangeSlider(
    value=[4E3, 2E9], min=1E3, max=2.5E9, step=1E3, 
    description='M_stars', layout=slider_layout, style={'description_width': '150px'}, readout_format='.0f'
)

m_dm_slider = widgets.FloatRangeSlider(
    value=[1E8, 5E10], min=1E8, max=6E10, step=1E8, 
    description='M_DM', layout=slider_layout, style={'description_width': '150px'}, readout_format='.0f'
)

sliders = [t_infall_slider, r_peri_slider, r_apo_slider, angle_slider, sigma_slider, r50_slider, m_stars_slider, m_dm_slider]

display(widgets.VBox(sliders))
```
